```python
import numpy as np
import matplotlib.pyplot as plt
```

# Problem A

```python
def inverse_distribution(n, given_lambda):
    return - np.log(1.0 - n) / given_lambda
```

```python
def generate_enqueue_and_completition_intervals(N, lambda_a, lambda_d):
    n = np.random.uniform(0, 1, N)
    c_t = inverse_distribution(n, lambda_d)
    q_t = inverse_distribution(n, lambda_a)
    return c_t, q_t
```

```python
class LabeledObject:
    label = "label"
    val = 0.0
    def __init__(self, label, val):
        self.val = val
        self.label = label
```

```python
def generate_arrival_and_delivery_times(N, c_t, q_t):
    A = [c_t[0]] 
    for i in range(1, N):
        A = np.append(A, A[i-1] + c_t[i])
    D = [c_t[0]+q_t[0]] 
    for i in range(1, N):
        D = np.append(D, max(D[i - 1], A[i]) + q_t[i])
    return A, D
```

```python
def generate_queue(AD):
    AD.sort(key=lambda x: x.val)
    queue = [0]
    counter=0
    ax = []
    for ad in AD:
        if ad.label == "A":
            queue.append(queue[-1] + 1)
            ax.append(counter)
        else:
            queue.append(queue[-1] - 1)
            counter+=1
    return queue, ax
```

```python
N = 10
lambda_a = 2.0
lambda_d = 2.5

c_t, q_t = generate_enqueue_and_completition_intervals(N, lambda_a, lambda_d)
A, D = generate_arrival_and_delivery_times(N, c_t, q_t)
A_mapped = list(map(lambda a: LabeledObject("A", a), A))
D_mapped = list(map(lambda d: LabeledObject("D", d), D))
AD = A_mapped + D_mapped
queue, ax = generate_queue(AD)

print(ax)

time = np.linspace(1, N, N)

print(A)
print(D)
scale_A = [100 * i for i in c_t]
scale_D = [100 * i for i in q_t]

plt.scatter(A, ax, label='Arriving tasks')
plt.scatter(D, range(1,len(D)+1), s=scale_D, label='Completed tasks')
plt.legend()
plt.show()
print()

plt.plot([0] + list(map(lambda a: a.val, AD)), queue, 'o', label='Queue')
plt.legend()
plt.show()
```

```python
Ri = list(map(lambda d,a: d-a, D,A ))
plt.scatter(D,Ri,label="Ri")
plt.legend()
plt.show()
```

# Problem B

```python
N = 10000
lambda_a = 2.0
lambda_d = 3.0

c_t, q_t = generate_enqueue_and_completition_intervals(N, lambda_a, lambda_d)
A, D = generate_arrival_and_delivery_times(N, c_t, q_t)
A_mapped = list(map(lambda a: LabeledObject("A", a), A))
D_mapped = list(map(lambda d: LabeledObject("D", d), D))
AD = A_mapped + D_mapped
queue, ax = generate_queue(AD)

en_sum = 0
t_sum = 0
counter = 0
for q, ad in zip(queue, AD):
    if counter != 0:
        en_sum += q*(ad.val-AD[counter-1].val)
        t_sum+=(ad.val-AD[counter-1].val)
    else:
        en_sum += q*ad.val
        t_sum+=ad.val
    counter +=1 
    
print(en_sum/t_sum)

er_sum = 0
for a, d in zip(A,D):
    er_sum += d-a

print(lambda_a*er_sum/N)
```
